#### Exploration 4) Examining possible relationships among popular cryptocurrencies (Ether, Iota, Litecoin and Ripple)

```python
import pandas as pd
import glob
import numpy as np

path = r'''C:\Users\Garrick\Documents\Springboard\Capstone Project 1\BTC_Datasets_vol2'''
filenames = list(glob.glob(path + "\*.csv"))
file = filenames[0]

crypt_ex = pd.read_csv(file, parse_dates=['Date'], index_col='Date')
crypt_ex.columns = [x+"_"+file[-10:-4] for x in crypt_ex.columns]
for file in filenames[1:]:
    load = pd.read_csv(file, parse_dates=['Date'], index_col='Date')
    load.columns = [x+"_"+file[-10:-4] for x in load.columns]
    crypt_ex = pd.merge(crypt_ex, load, left_index=True, right_index=True)

crypt_ex.info()
```

Since we are given bid/ask data, I also wish to explore daily spread within each coin's market.  In capital markets, the spread is the value difference between a bidder's price and the seller's ask.  Theoretically, informationally efficient and liquid markets should see very little spread.  As a result, some will analyze spread as a proxy for volatility (higher spread can indicate dissonance in the market).  

```python
# creating new columns for daily inter-coin spread
crypt_ex['Spread_BTC'] = crypt_ex.Ask_BTCUSD.sub(crypt_ex.Bid_BTCUSD)
crypt_ex['Spread_ETH'] = crypt_ex.Ask_ETHUSD.sub(crypt_ex.Bid_ETHUSD)
crypt_ex['Spread_IOT'] = crypt_ex.Ask_IOTUSD.sub(crypt_ex.Bid_IOTUSD)
crypt_ex['Spread_LTC'] = crypt_ex.Ask_LTCUSD.sub(crypt_ex.Bid_LTCUSD)
crypt_ex['Spread_RIP'] = crypt_ex.Ask_XRPUSD.sub(crypt_ex.Bid_XRPUSD)
```

```python
pd.plotting.scatter_matrix(crypt_ex[['Spread_BTC','Spread_ETH','Spread_IOT','Spread_LTC','Spread_RIP']], diagonal="kde",
                            figsize=(20,20), alpha=1)
plt.tight_layout()
plt.show()
```

#### Exploration 3) Analysis:

Through our visual exploration of our data, it seems there may be some relationships between price and volume of coins (essentially intra-market dynamics).   However, the inter-coin spread does not appear to affect the other. This can be interpreted as, the volatility and market price risk in one coin does not appear to incite of affect price volatility in others.

The underlying curiosity I had set out to explore is whether we can predict Bitcoin prices based on the activity in other coins.  At this point, I wish to employ recently learned concepts in Machine Learning using scikit-learn on this coin data.  I want to see if it is possible to predict Bitcoin prices or not.  The outcome of either case would be insightful.

In this case, I will be using SciKit-Learn's Lasso Linear Regression as I do not know which features to include and to what degree.  Therefore, will be relying on Regularization via Lasso. 

Furthermore, I wish to apply a scaling method here to account for the different units of scale for each feature. 

```python
# importing essential libraries for modeling

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# creating inputs necessary for the models

col_list = ['Mid_ETHUSD','Mid_IOTUSD','Mid_LTCUSD','Mid_XRPUSD', 'Volume_BTCUSD',
            'Volume_ETHUSD','Volume_IOTUSD','Volume_LTCUSD','Volume_XRPUSD']  #omitting MID_BTC as it is our target

X = crypt_ex[col_list]
y = crypt_ex.Mid_BTCUSD.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=16)
```

```python
# instantiating and fitting the model

scaler = StandardScaler()
lasso  = Lasso(alpha=0.6)

pipeline = make_pipeline(scaler, lasso)
lasso_scaled = pipeline.fit(X_train, y_train)
```

```python
# plotting the coefs of the features computed by the lasso model

lasso_coef = lasso.coef_

plt.plot(range(len(X.columns)), lasso_coef)
plt.xticks(range(len(X.columns)), X.columns.values, rotation=60)
plt.margins(0.02)
plt.show()
print(lasso_coef)
```

```python
lasso_scaled.score(X_test, y_test)
```

```python
x_new = np.array([[1115.85, 2.2765, 163.35, 1.1269, 43450, 136162, 6582122, 145831 , 68250520]]) # jan 31 data

# yesterday's BTC mid price was 10,192

lasso_scaled.predict(x_new)
```

```python
x_new2 = np.array([[981.91, 2.5098, 177.83, 1.3299, 63857, 227914, 13140109, 204099 , 162534798]]) # Jan 24 data

# jan 24 BTC mid price = 10,827
lasso_scaled.predict(x_new2)
```

According to the Lasso model, the price of Litecoin, Ripple, and Ethereum volumes are the most significant features in predicting the price of Bitcoin, Y.  The model also has an accuracy of 89.4% given the test data.  However, the predict values come relatively close (in the grand scheme) but not quite on the mark.  The prediction dates are only separated by a week, but I found that subtracting ~$3,500 off the prediction I can get quite close.  

Let's try as a normal linear regression since we only have limited features. 

```python
linreg = LinearRegression()

pipeline2 = make_pipeline(scaler, linreg)

linreg_scaled = pipeline2.fit(X_train, y_train)
```

```python
linreg_scaled.score(X_test, y_test)
```

```python
lin_coef = linreg.coef_

plt.plot(range(len(X.columns)), lin_coef)
plt.xticks(range(len(X.columns)), X.columns.values, rotation=60)
plt.margins(0.02)
plt.xlabel('Covariates')
plt.ylabel('Weighting')
plt.show()
print(lin_coef)
```

```python
linreg_scaled.predict(x_new)
# yesterday's BTC mid price was 10,192
```

```python
# jan 24 BTC mid price = 10,827
linreg_scaled.predict(x_new2)
```

Given the similar correlation proportions as in the Lasso model, it is unsurprising that the regular Linear Model performs similarly to the Lasso.  

In both cases, the model is able to acheive an 89% accuracy on the testing data.  However, they appear to really miss the mark in the newer unseen data.  I noticed that subtracing 3,500 from the prediction brings the prediction closer to the actual value.  This may indicate that the model is missing one or more features which may help it improve.
