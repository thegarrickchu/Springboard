Notes: Examining possible relationships among popular cryptocurrencies (namely Bitcoin, Ether, Iota, Litecoin and Ripple)

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

path = r'''C:\Users\Garrick\Documents\Springboard\Capstone Project 1\BTC_Datasets_vol2'''
filenames = list(glob.glob(path + "\*.csv"))
file = filenames[0]

crypt_ex = pd.read_csv(file, parse_dates=['Date'], index_col='Date')
crypt_ex.columns = [x+"_"+file[-10:-4] for x in crypt_ex.columns]
for file in filenames[1:]:
    load = pd.read_csv(file, parse_dates=['Date'], index_col='Date')
    load.columns = [x+"_"+file[-10:-4] for x in load.columns]
    crypt_ex = pd.merge(crypt_ex, load, left_index=True, right_index=True)

crypt_ex.info()
```

1/11/18 Data Dictionary:

All datasets are for that cryptocurrency in USD and includes daily low, high, bid, ask and volume.

BTC = Bitcoin

ETH = Ethereum

IOT = Iota

LTC = Litecoin

XRP = Ripple

Iota was launched in June 2017 and thus the dataframe only includes data starting for when Iota was introduced.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

crypt_ex.plot(subplots=True, figsize=(20,100))
plt.show()
```

```python
#using the "Mid" series for plotting as proxy for daily crypto prices

pd.plotting.scatter_matrix(crypt_ex[['Mid_BTCUSD','Mid_ETHUSD','Mid_IOTUSD','Mid_LTCUSD','Mid_XRPUSD']], diagonal="kde",
                            figsize=(20,20), alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#is there any visual evidence of relationship between cryptovolumes on a large scale?
pd.plotting.scatter_matrix(crypt_ex[['Volume_BTCUSD','Volume_ETHUSD','Volume_IOTUSD','Volume_LTCUSD','Mid_XRPUSD']], diagonal="kde",
                            figsize=(20,20), alpha=1)
plt.tight_layout()
plt.show()
```

```python
crypt_ex['Spread_BTC'] = crypt_ex.Ask_BTCUSD.sub(crypt_ex.Bid_BTCUSD)
crypt_ex['Spread_ETH'] = crypt_ex.Ask_ETHUSD.sub(crypt_ex.Bid_ETHUSD)
crypt_ex['Spread_IOT'] = crypt_ex.Ask_IOTUSD.sub(crypt_ex.Bid_IOTUSD)
crypt_ex['Spread_LTC'] = crypt_ex.Ask_LTCUSD.sub(crypt_ex.Bid_LTCUSD)
crypt_ex['Spread_RIP'] = crypt_ex.Ask_XRPUSD.sub(crypt_ex.Bid_XRPUSD)
#spread is often a proxy to volatility. spread refers to the value difference between a sellers asking price and a buyers bid
#large spreads typically indicate that dissonance in market sentiment
```

```python
pd.plotting.scatter_matrix(crypt_ex[['Spread_BTC','Spread_ETH','Spread_IOT','Spread_LTC','Spread_RIP']], diagonal="kde",
                            figsize=(20,20), alpha=1)
plt.tight_layout()
plt.show()
```
